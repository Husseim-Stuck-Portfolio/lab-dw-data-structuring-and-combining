## Challenge 1: Combining & Cleaning Data

In this challenge, we will be working with the customer data from an insurance company, as we did in the two previous labs. The data can be found here:
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv

But this time, we got new data, which can be found in the following 2 CSV files located at the links below.

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv

Note that you'll need to clean and format the new data.

Observation:
- One option is to first combine the three datasets and then apply the cleaning function to the new combined dataset
- Another option would be to read the clean file you saved in the previous lab, and just clean the two new files and concatenate the three clean datasets

```python
# Your code goes here

import pandas as pd
df1 =  pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv")
df2 =  pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv")
df3 =  pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv")

print("Shapes:", df1.shape, df2.shape, df3.shape)
```

```python
print("File1 Columns:", df1.columns.tolist())
print("\n Missing Values in file1:")
print(df1.isnull().sum())
print("\n Data type:")
print(df1.dtypes)
```

```python
# Cleaning Function for all files
def clean_df(df):
    #standarize spaces, casing
    df.columns = df.columns.str.strip().str.replace(" ","_").str.lower()
    #clean number colums
    numeric_cols = ["income", "monthly_premium_auto", "total_claim_amount"]
    for col in numeric_cols:
            if col in df.columns:
                df[col] = pd.to_numeric(df[col].astype(str).str.replace(",", ""), errors = "coerce")
    #then cleaning the text columns
    text_cols = ["state", "gender", "education", "vehicle_class", "sales_channel"]
    for col in text_cols:
            if col in df.columns:
                df[col] = df[col].astype(str).str.strip().str.title()
    df = df.fillna("unknown")
    df = df.drop_duplicates()

    return df
```

```python
df1_clean = clean_df(df1)
df2_clean = clean_df(df2)
df3_clean = clean_df(df3)

print("Clean shapes:", df1_clean.shape, df2_clean.shape, df3_clean.shape)
```

```python
df_combined = pd.concat([df1_clean, df2_clean, df3_clean], ignore_index=True)
print("Combined shape:", df_combined.shape)
print(df_combined.head())
```
